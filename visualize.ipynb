```python
# 3x3 그리드로 각 (6, 6) 배열 시각화
score = np.load(os.path.join(path, "score_board.npy"))
score_viz = score[:, :, :, :, 0]

fig, axes = plt.subplots(3, 3, figsize=(12, 12))

for i in range(3):
    for j in range(3):
        ax = axes[i, j]
        ax.imshow(score_viz[i, j], cmap="seismic")
        ax.set_title(f"Subplot [{i},{j}]")
        ax.axis("off")  # 축 표시 제거

plt.tight_layout()  # 서브플롯 간격 조정
plt.show()
```
